### 书稿中的代码是在TF1.0版本下，现在主流平台已更新到TF2.0，因此代码有微调

```python
#代码示例：4-1
import tensorflow as tf
# [旧写法] tf.compat.v1.disable_eager_execution()  # TF1 风格，需要手动禁用 eager 模式
# ↑ TF 2.0+ 默认开启 eager execution，不再需要 Session
X=tf.constant(
    [
        [
            [[3],[-1],[3]],
            [[9],[2],[4]],
            [[8],[1],[5]]
        ]
    ]
    ,tf.float32
)
K=tf.constant(
    [
        [
            [[5]],[[2]]],
        [
            [[1]],[[3]]
        ]
    ]
    ,tf.float32
)
# same卷积
conv=tf.nn.conv2d(X,K,(1,1,1,1),'SAME')
# [旧写法] session=tf.compat.v1.Session()
# [旧写法] print(session.run(conv))
# ↑ TF1 的 Session 在 TF 2.0+ 中已不推荐使用
# [新写法] 适用于 TensorFlow >= 2.0，直接使用 eager execution
print(conv.numpy())
```
