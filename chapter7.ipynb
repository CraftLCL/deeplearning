```python
#代码示例：7-7
# 使用pandas读入数据
import pandas as pd
poems_text = pd.read_table('./poems_clean.txt', header=None)
poems_text.columns = ["text"]
# 查看文本
poems_text.head()
```

```python
#代码示例：7-8
import string
import numpy as np
poems_new = []
for line in poems_text['text']:  # poems_text[0]的第0列，指
    title, poem = line.split(':')
    poem = poem.replace(' ', '') #将空格去掉
    poem = 'bbb' + poem
    poems_new.append(list(poem))
```

```python
#代码示例：7-9
XY  =[]
for poem in poems_new:
    for i in range(len(poem) - 3):
        x1 = poem[i]
        x2 = poem[i+1]
        x3 = poem[i+2]
        
        y = poem[i+3]
        XY.append([x1, x2, x3, y])
```

```python
# 展示整理后的X和Y的形式
print("原始诗句：")
print(poems_text['text'][3864])
print("\n")
print("训练数据：")
print(["X1", "X2", "X3", "Y"])
for i in range(132763, 132773):
    print(XY[i])
```

```python
#代码示例7-10
# [旧写法] from keras.preprocessing.text import Tokenizer
# ↑ 独立 keras 包在 TF 2.0+ 中已不推荐使用
# [新写法] 适用于 TensorFlow >= 2.0
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(poems_new)
print(tokenizer.word_index)

#tokenizer默认把0这个索引留个停止词了，其它的词是从1开始索引的。
#然后分类的损失又默认0是第一类
#所以导致假设我们有三个字：a，b，c。
#他们的索引分别是1，2，3，没有0。
#这对keras要使用一个4分类才行
vocab_size = len(tokenizer.word_index) + 1
```

```python
#代码示例：7-11
XY_digit = np.array(tokenizer.texts_to_sequences(XY))
X_digit = XY_digit[:, :3]
Y_digit = XY_digit[:, 3]
for i in range(132763, 132773):
    print("{:<35}".format(str(XY[i])), "\t", "{:<30}".format(str(list(X_digit[i]))),"\t", Y_digit[i])
```

```python
#代码示例：7-12
# Embedding + 线性模型
# [旧写法] from keras.layers import Input, Embedding
# [旧写法] from keras.models import Sequential, load_model, Model
# [旧写法] from keras.layers import Input, Dense, Activation, Embedding, Flatten
# [新写法] 适用于 TensorFlow >= 2.0
from tensorflow.keras.layers import Input, Embedding
from tensorflow.keras.models import Sequential, load_model, Model
from tensorflow.keras.layers import Dense, Activation, Flatten
hidden_size = 256

inp = Input(shape=(3,))
x = Embedding(vocab_size, hidden_size)(inp)
x = Flatten()(x)
x = Dense(vocab_size)(x)
pred = Activation('softmax')(x)

model = Model(inp, pred)
model.summary()
```

```python
#代码示例：7-13
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_digit,Y_digit,test_size=0.2, random_state=0)

# [旧写法] from keras.optimizers import Adam
# [新写法] 适用于 TensorFlow >= 2.0
from tensorflow.keras.optimizers import Adam
# [旧写法] model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(lr=0.001))
# [新写法] 适用于 TensorFlow >= 2.11
model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(learning_rate=0.001))
model.fit(X_train, Y_train, validation_data=(X_test, Y_test), batch_size=10000, epochs=10)
```

```python
#代码示例：7-32
import string
import jieba

English = []
Chinese = []

f = open('./cmn.txt', "r", encoding='utf-8')
for line in f.readlines():
    # In each line, the data is of format:
    #    He ran.	他跑了。
    eng, chs = line.strip().split('\t')
    
    eng = eng[:-1]
    eng = eng.split(' ')
    English.append(eng)
    
    #分词
    chs = jieba.lcut(chs)
    chs = ['B'] + chs
    Chinese.append(chs)

# Example Print
print(English[20000])
print(Chinese[20000])
```

```python
#代码示例：7-33
# [旧写法] from keras.preprocessing.text import Tokenizer
# [旧写法] from keras.preprocessing.sequence import pad_sequences
# [新写法] 适用于 TensorFlow >= 2.0
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Process English Data
tokenizer_eng = Tokenizer()
tokenizer_eng.fit_on_texts(English)
eng_vocab_size = len(tokenizer_eng.word_index) + 1

# Process Chinese Data
tokenizer_chs = Tokenizer()
tokenizer_chs.fit_on_texts(Chinese)
chs_vocab_size = len(tokenizer_chs.word_index) + 1

print("英文字典的大小,",eng_vocab_size)
print("中文字典的大小,",chs_vocab_size)
```

```python
#代码示例：7-34
#补全英文样本
max_length=40
eng_digit = tokenizer_eng.texts_to_sequences(English)
eng_digit = pad_sequences(eng_digit, maxlen=max_length, padding='post')

#补全中文样本
chs_digit = tokenizer_chs.texts_to_sequences(Chinese)
chs_digit = pad_sequences(chs_digit, maxlen=max_length, padding='post')

print("展示补全后的数据维度")
print(eng_digit.shape)
print(chs_digit.shape)
print("\n")

print("展示第6条样本")
print(English[6])
print(Chinese[6])
print("\n")

print("展示处理后的结果")
print(eng_digit[6])
print(chs_digit[6])
```

```python
#代码示例：7-35
X_eng = eng_digit
X_chs = chs_digit[:, :-1]
Y = chs_digit[:, 1:]
print(X_eng.shape)
print(X_chs.shape)
print(Y.shape)
```

```python
#代码示例：7-36
# [旧写法] from keras.utils import to_categorical
# [新写法] 适用于 TensorFlow >= 2.0
from tensorflow.keras.utils import to_categorical
Y = to_categorical(Y, num_classes=chs_vocab_size)
print(Y.shape)
```

```python
#代码示例：7-37
# [旧写法] from keras.layers import Input, LSTM, Dense, Embedding
# [旧写法] from keras.models import Model, load_model
# [新写法] 适用于 TensorFlow >= 2.0
from tensorflow.keras.layers import Input, LSTM, Dense, Embedding
from tensorflow.keras.models import Model, load_model
hidden_size_1 = 64
hidden_size_2 = 128

encoder_input = Input(shape = (max_length,))
x = Embedding(eng_vocab_size, hidden_size_1, mask_zero=True, name='ENG')(encoder_input)
encoder_output,encoder_h,encoder_c = LSTM(hidden_size_2, return_state=True, name='Encoder')(x)

encoder_state = [encoder_h,encoder_c]
```

```python
#代码示例：7-38
decoder_input = Input(shape=(max_length-1, ))
x = Embedding(chs_vocab_size, hidden_size_1, mask_zero=True, name='CHS')(decoder_input)
decoder_output = LSTM(hidden_size_2, return_sequences=True, name='Decoder')(x, initial_state=encoder_state)

pred = Dense(chs_vocab_size, activation='softmax')(decoder_output)
```

```python
#代码示例：7-39
model = Model([encoder_input,decoder_input],pred)
model.summary()
chs_vocab_size
```

```python
#代码示例：7-40
# [旧写法] from keras.optimizers import Adam
# [新写法] 适用于 TensorFlow >= 2.0
from tensorflow.keras.optimizers import Adam
# [旧写法] model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=0.01), metrics=['accuracy'])
# [新写法] 适用于 TensorFlow >= 2.11
model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.01), metrics=['accuracy'])
model.fit([X_eng, X_chs], Y, epochs=10, batch_size=512, validation_split=0.2)
```
